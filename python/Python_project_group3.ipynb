## 2A.7 Outlier Detection and Treatment

To systematically evaluate outliers across all numeric variables, we apply the **Interquartile Range (IQR) method**.  
For each variable, we compute:

- **Lower bound** = Q1 − 1.5 × IQR  
- **Upper bound** = Q3 + 1.5 × IQR  
- **Left outliers** = values < lower bound  
- **Right outliers** = values > upper bound  
- **Total outliers** = left + right  
- **Outlier percentage** = total outliers / total observations  

We use a **5% threshold** to determine whether outliers are acceptable or require further investigation.


### Interpretation

The IQR-based outlier summary allows us to evaluate each numeric variable objectively.  
In financial datasets, extreme values often represent **legitimate borrower behavior**, not data errors.  

Using the 5% threshold:

- Variables with **< 5% outliers** are retained without modification  
- Variables with **> 5% outliers** are flagged for review, but not automatically capped or removed  

This approach ensures that we preserve meaningful variability while avoiding unnecessary distortion of the data.

```python
### Outlier Detection Function (IQR Method)
import pandas as pd
import numpy as np

def iqr_outlier_summary(df, columns, threshold=0.05):
    """
    Computes IQR-based outlier statistics for a list of numeric columns.
    Returns a summary DataFrame with:
    - left outliers
    - right outliers
    - total outliers
    - percentage
    - decision (below/above threshold)
    """

    results = []

    for col in columns:
        series = df[col].dropna()

        Q1 = series.quantile(0.25)
        Q3 = series.quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        left = (series < lower).sum()
        right = (series > upper).sum()
        total = left + right
        pct = total / len(series)

        decision = "Below threshold (keep)" if pct <= threshold else "Above threshold (review)"

        results.append({
            "Variable": col,
            "Left_Outliers": left,
            "Right_Outliers": right,
            "Total_Outliers": total,
            "Outlier_%": round(pct * 100, 2),
            "Decision": decision
        })

    return pd.DataFrame(results)
```

```python
### Generate Outlier Summary Table
numeric_cols = [
        'requested_amount_usd', 'monthly_installment_usd',
       'last_payment_amount_usd', 'annual_income_usd', 'debt_to_income_ratio',
       'credit_score_low', 'credit_score_high', 'open_credit_lines',
       'public_records_count', 'interest_rate', 'revolving_utilization'
       ]

outlier_table = iqr_outlier_summary(apps, numeric_cols, threshold=0.05)
outlier_table
```

any variable that directly encodes post‑outcome information or reflects borrower behavior after loan origination must be excluded from the modeling dataset.

## 1. Identify variables that leak information about the target

A leakage variable is any feature that:

- is created after the loan is issued  
- reflects repayment behavior that occurs after origination  
- directly encodes the final loan outcome  
- or is logically dependent on the target variable  

Based on the dataset structure and domain logic, the following variables are leakage risks:

### outcome_status
This variable explicitly states whether the loan was fully paid or charged off. Since target_default is derived directly from it, keeping it would give the model the answer.

Decision: Remove.

### last_payment_amount_usd
This value reflects borrower payments made after the loan was issued. Borrowers who default typically have low or zero last payments, while fully paid loans show large final payments. This variable therefore contains post‑outcome behavior.

Decision: Remove.

---

The following variables were evaluated for leakage but determined to be safe because they reflect borrower characteristics or credit bureau information available at the time of application:

- requested_amount_usd

- loan_purpose  
- region_code  
- debt_to_income_ratio  
- credit_score_low  
- credit_score_high  
- open_credit_lines  
- public_records_count  
- interest_rate  
- revolving_utilization  

These variables represent information known at origination and do not reveal the final loan outcome.

Decision: Retain.

---

## 3. Final list of leakage variables removed

The following variables were removed from the modeling dataset:

- outcome_status  
- last_payment_amount_usd  

```python
# To revome the leakage variables
apps.drop(columns=['outcome_status', 'last_payment_amount_usd'], inplace=True)
```

## 4. Resulting dataset

After removing leakage variables, the dataset now contains only features available at the time of loan application. This ensures that the predictive model does not gain unfair advantage from information that would not be available in a real‑world lending scenario.
