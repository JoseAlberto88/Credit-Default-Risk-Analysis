# 1. Project Overview

This project examines credit default risk using two complementary datasets: a loan applications dataset used for predictive modeling, and a credit inquiries dataset used for business insights. Together, these datasets provide a comprehensive view of borrower behavior, credit demand, and repayment outcomes.

---

## Applications Dataset. Used for Predictive Modeling

Each row in the applications dataset represents a completed loan application with detailed borrower, loan, and credit profile information.  
The dataset includes the following key fields:

- application_id — Unique identifier for each loan application  
- requested_amount_usd — Loan amount requested  
- repayment_term_text — Repayment duration as text  
- interest_rate_text — Interest rate as text  
- monthly_installment_usd — Scheduled monthly payment  
- last_payment_amount_usd — Most recent payment amount  
- risk_grade_band and risk_grade_detail — Credit risk categories  
- employment_tenure_text — Borrower employment length  
- housing_status — Housing situation  
- annual_income_usd — Reported annual income  
- income_verification_status — Verification category  
- origination_month — Month the loan originated  
- outcome_status — Final loan outcome  
- loan_purpose — Borrower-stated purpose  
- region_code — Geographic region  
- debt_to_income_ratio — Borrower DTI  
- credit_score_low and credit_score_high — Credit score range  
- open_credit_lines — Number of open accounts  
- public_records_count — Number of derogatory public records  
- revolving_utilization_text — Revolving credit utilization  

Because this dataset contains the final loan outcome, it is used to build the predictive model.

---

## Prediction Target

The goal of the predictive component is to determine whether a loan application results in default.  
The target variable is defined as:

- target_default = 1 if outcome_status == "Charged Off"  
- target_default = 0 otherwise  

This converts the problem into a binary classification task suitable for machine learning.

---

## Inquiries Dataset. Used for Business Insights

The inquiries dataset contains information about credit inquiries submitted by potential borrowers.  
It includes the following fields:

- requested_amount_usd — Amount requested during inquiry  
- application_date — Date of inquiry  
- request_title — Free-text description of the request  
- screening_score — Preliminary risk score  
- debt_to_income_text — DTI expressed as text  
- region_code — Geographic region  
- employment_tenure_text — Employment length  
- policy_code — Policy indicator  

This dataset does not contain loan outcomes and therefore cannot be used for supervised learning.  
Instead, it supports business insights by revealing patterns in borrower demand, screening score distributions, regional inquiry trends, and employment characteristics at inquiry time.

---

## Model Evaluation Approach


The model will be evaluated using three complementary methods:

1. **Confusion Matrix**  
   Used to understand the distribution of correct and incorrect predictions.

2. **A false-positive–focused metric**  
   This directly supports the business objective of minimizing cases where the model incorrectly predicts that a borrower will repay when they actually default.

3. **ROC–AUC Curve**  
   Used to assess overall model discrimination and to help analyze potential overfitting or underfitting.
